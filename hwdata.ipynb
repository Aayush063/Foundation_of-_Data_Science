```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("Health_heart_experimental.csv")
df.head(10)
```

LINE CHART


```python
age = [21,30,35,39,50]
bmi = [21,21,20,33,19]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(age, bmi, color='blue', marker='+', linestyle='solid')

# add a title
plt.title(" Health check bmi per age")

# add a label to the y-axisYears
plt.ylabel("BMI")
plt.xlabel("age")
plt.show()
```

From the above line chart we can analyze that the person till 35 are in the normal condition  then after aged are prone to more heart diseases and diabetes due to increase in the BNI.

BAR PLOT

```python
avg_bmi_by_sex = df.groupby('sex')['BMI'].mean()

print(avg_bmi_by_sex)

# Bar plot
avg_bmi_by_sex.plot(kind='bar')
plt.xlabel('Sex (0 = Female, 1 = Male)')   # adjust labels if needed
plt.ylabel('Average BMI')
plt.title('Average BMI by Sex')
plt.xticks(rotation=0)
plt.show()
```

From the above bar plot we found out that the AVG BMI was seen almost equal  in SEX"0" and  in SEX"1" which concludes us to point that the both sex are prone to heart disease.

PIE CHART

```python
sex_counts = df['sex'].value_counts().sort_index()

labels = ['Female', 'Male']

plt.figure()
plt.pie(sex_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution (Male vs Female)')
plt.axis('equal')   
```

From the above pie chart we found that the survey was taken in almost equal number of male and female gender

BOX PLOT

```python
plt.figure()
df[['SysBP', 'DiaBP']].boxplot()
plt.ylabel('Blood Pressure (mmHg)')
plt.title('Blood Pressure Box Plot')
plt.show()
```

The above box plot shows about the variation of systolic and diastolic blood pressure and range of them. We can see that the syslotic blood pressure is in range from 115 to 135 and diastolic pressure is from 85 to 108

HISTOGRAM

```python
plt.figure()
plt.hist(df['HR'], bins=10)
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Frequency')
plt.title('Histogram of Heart Rate')
plt.show()
```

The following diagram shows that the bars are uniform in height which shows that they are evenly distirbuted in the range.

SCATTER PLOT

```python
df_500 = df.head(500)

plt.figure()
plt.scatter(
    df_500['HR'],
    df_500['SysBP'],
    c=df_500['BMI'],
    s=df_500['age'],
)

plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Systolic Blood Pressure (mmHg)')
plt.title('HR vs SysBP (Color = BMI, Size = Age) [500 Records]')
plt.colorbar(label='BMI')
plt.show()
```

the following shows the scatter plot of 4 different variable which are systolic blood pressure, heart rate and BMI represented by colour gradient of less by dark blue and more by yellow. similarly the age is represented by size of the point. small point shows small age whereas big point shows big age. 
It shows that the higher blood pressure have comparatively less BMI and vice cersa.
